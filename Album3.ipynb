```python
import pandas as pd

# Wczytaj dane z pliku Excel
df = pd.read_excel('Album.xlsx')

# Zmień nazwy kolumn
df = df.rename(columns={'POS': 'POZYCJA' , 'TITLE': 'TYTUŁ', 'ARTIST': 'ARTYSTA', 'YEAR': 'ROK'})

# Zamień nagłówki na małe litery z wyjątkiem pierwszej litery
df.columns = df.columns.str.title()

# Usuń kolumnę 'High Posn'
df = df.drop(columns=['High Posn'])

# Znajdź unikalnych artystów
ilosc_artystow = df['Artysta'].nunique()
print(f'Ilość pojedynczych artystów: {ilosc_artystow}')

# Znajdź najczęściej pojawiające się zespoły
najczestsze_zespoly = df['Artysta'].value_counts().idxmax()
print(f'Najczęściej pojawiający się zespół: {najczestsze_zespoly}')

# Znajdź rok z największą ilością albumów
najwiecej_albumow_rok = df['Rok'].value_counts().idxmax()
print(f'Rok z największą liczbą albumów: {najwiecej_albumow_rok}')

# Znajdź albumy wydane między 1960 a 1990 rokiem
albumy_1960_1990 = df[(df['Rok'] >= 1960) & (df['Rok'] <= 1990)]
print(f'Liczba albumów wydanych między 1960 a 1990 rokiem (włącznie): {len(albumy_1960_1990)}')

# Znajdź najmłodszy album
najmlodszy_album_rok = df['Rok'].max()
print(f'Najmłodszy album został wydany w roku: {najmlodszy_album_rok}')

# Znajdź najwcześniej wydane albumy każdego artysty
najwczesniejsze_albumy = df.groupby('Artysta')['Rok'].min().reset_index()
print('\nNajwcześniej wydane albumy każdego artysty:')
print(najwczesniejsze_albumy)

# Ustaw nazwę pliku do zapisu w formacie CSV
nazwa_pliku = 'najwczesniejsze_albumy.csv'

# Zapisz dane do pliku CSV
df.to_csv(nazwa_pliku, index=False)

# Dodaj wyniki najwczesniejszych albumow
najwczesniejsze_albumy.to_csv(nazwa_pliku, mode='a', header=False, index=False)
```
